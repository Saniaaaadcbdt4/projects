*1. Data Set Selection*
Use the "Palmer Penguins" dataset from the seaborn library.


```python
# 2. Data loading
import pandas as pd
import seaborn as sns

# Load the dataset
penguins = sns.load_dataset('penguins')

# Display the first few rows of the dataset
penguins.head()
```

```python
# 3. Data exploration
# Display the structure of the dataset
penguins.info()

# Display the statistical summary of the dataset
penguins.describe()

# Display the summary of categorical features
penguins.describe(include='O')
```

```python
# 4. Data-cleaning
# Check for missing values
missing_values = penguins.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Handle missing values (for simplicity, let's drop rows with missing values in this example)
penguins_cleaned = penguins.dropna()

# Remove duplicates
penguins_cleaned = penguins_cleaned.drop_duplicates()

# Check for missing values and duplicates after cleaning
print("Missing values after cleaning:\n", penguins_cleaned.isnull().sum())
print("Number of duplicates after cleaning:", penguins_cleaned.duplicated().sum())
```

```python
# 5. Data-visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Visualization 1: Distribution of penguins by species
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=penguins_cleaned, palette='muted')
plt.title('Distribution of Penguins by Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
# Insight: This plot shows the number of penguins in each species.

# Visualization 2: Flipper length distribution by species
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='flipper_length_mm', data=penguins_cleaned, palette='muted')
plt.title('Flipper Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.show()
# Insight: This plot shows the flipper length distribution for each species.

# Visualization 3: Bill length vs. bill depth
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', hue='species', data=penguins_cleaned, palette='muted')
plt.title('Bill Length vs. Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.show()
# Insight: This plot shows the relationship between bill length and bill depth for different species.

# Visualization 4: Body mass distribution by species and sex
plt.figure(figsize=(10,6))
```

```python

sns.boxplot(x='species', y='body_mass_g', hue='sex', data=penguins_cleaned, palette='muted')
plt.title('Body Mass Distribution by Species and Sex')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
# Insight: This plot shows the body mass distribution for each species, separated by sex.

# Visualization 5: Pairplot of numerical features
sns.pairplot(penguins_cleaned, hue='species', palette='muted', diag_kind='kde')
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
# Insight: This pairplot shows relationships between numerical features, separated by species, which helps in identifying patterns and differences among species.
```

6. **Analysis and Insights**
Distribution of Penguins by Species: This plot shows the number of penguins in each species, indicating the relative population sizes in the dataset.

**Flipper Length Distribution by Species**: The boxplot shows the distribution of flipper lengths for each species, highlighting the differences in flipper lengths among species.

**Bill Length vs. Bill Depth**: The scatter plot shows the relationship between bill length and bill depth, separated by species. It highlights distinct clusters for each species.

**Body Mass Distribution by Species and Sex:** The boxplot shows the body mass distribution for each species, separated by sex, indicating differences in body mass between males and females within each species.

**Pairplot of Numerical Features:** The pairplot provides a comprehensive view of relationships between numerical features, separated by species. It helps in identifying patterns and differences among species in terms of bill length, bill depth, flipper length, and body mass.
